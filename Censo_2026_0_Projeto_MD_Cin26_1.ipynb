```python
import pandas as pd
import numpy as np

# 1. Carregar a base (ajuste o nome do arquivo para o seu)
# Supondo que o arquivo esteja na raiz do Colab
df_bruto = pd.read_csv('/content/base_unificada.csv', sep=',') # ou o delimitador correto

# 2. Filtrar apenas Licenciaturas (Isolamento de Domínio)
df_licenciaturas = df_bruto[df_bruto['tipo_grau_academico'] == 'Licenciatura'].copy()

# 3. Seleção das colunas-mestre (As 18 colunas que listamos)
colunas_selecionadas = [
    'ano', 'id_curso', 'nome_area_detalhada', 'rede', 'tipo_modalidade_ensino',
    'quantidade_vagas', 'quantidade_inscritos', 'quantidade_ingressantes',
    'quantidade_matriculas', 'quantidade_concluintes',
    'quantidade_alunos_situacao_trancada', 'quantidade_alunos_situacao_desvinculada',
    'quantidade_matriculas_financiamento_reembolsavel_fies',
    'quantidade_matriculas_financiamento_nao_reembolsavel_prouni_integral',
    'quantidade_matriculas_parfor', 'quantidade_alunos_apoio_social', 'id_municipio'
]

df_estudo = df_licenciaturas[colunas_selecionadas].copy()

# 4. Ordenação para a Janela Deslizante (Fundamental)
df_estudo = df_estudo.sort_values(['id_curso', 'ano'])

# 5. Engenharia de Atributos Automática (O "Coração" do Projeto)
# Vamos criar os Lags (dados de 3 anos atrás)
df_estudo['matriculas_lag3'] = df_estudo.groupby('id_curso')['quantidade_matriculas'].shift(3)
df_estudo['ingressantes_lag1'] = df_estudo.groupby('id_curso')['quantidade_ingressantes'].shift(1)

# Criando os Deltas (As variáveis que a árvore vai usar)
# Tendência de Matrícula (3 anos)
df_estudo['delta_matricula_3anos'] = (df_estudo['quantidade_matriculas'] - df_estudo['matriculas_lag3']) / df_estudo['matriculas_lag3']

# Taxa de Ocupação de Vagas
df_estudo['taxa_ocupacao'] = df_estudo['quantidade_ingressantes'] / df_estudo['quantidade_vagas']

# Índice de Evasão (Desvinculados / Matrículas)
df_estudo['indice_evasao'] = df_estudo['quantidade_alunos_situacao_desvinculada'] / df_estudo['quantidade_matriculas']

# Dependência de Fomento (Fies + Prouni)
df_estudo['dep_fomento'] = (df_estudo['quantidade_matriculas_financiamento_reembolsavel_fies'] +
                            df_estudo['quantidade_matriculas_financiamento_nao_reembolsavel_prouni_integral']) / df_estudo['quantidade_matriculas']

# 6. Criando a Target (O Alvo: Risco de Colapso)
# Definimos Risco (1) se as matrículas caíram mais de 50% ou o curso zerou no futuro (T+2)
df_estudo['matriculas_futuro'] = df_estudo.groupby('id_curso')['quantidade_matriculas'].shift(-2)
df_estudo['TARGET_RISCO'] = np.where((df_estudo['matriculas_futuro'] < (df_estudo['quantidade_matriculas'] * 0.5)) |
                                      (df_estudo['matriculas_futuro'].isna()), 1, 0)

# 7. Limpeza Final
# Removemos as linhas onde não foi possível calcular o passado (os primeiros 3 anos de cada curso)
df_final = df_estudo.dropna(subset=['delta_matricula_3anos', 'matriculas_futuro'])

# 8. Exportar o novo Dataset para exploração
df_final.to_csv('dataset_modelagem_licenciaturas.csv', index=False)

print("Novo dataset gerado com sucesso! Pronto para a Árvore de Decisão.")
```

```python
import pandas as pd
import numpy as np

# 1. Definir as colunas indispensáveis (reduzindo a carga inicial)
colunas_foco = [
    'ano', 'id_curso', 'tipo_grau_academico', 'nome_area_detalhada',
    'quantidade_vagas', 'quantidade_ingressantes', 'quantidade_matriculas',
    'quantidade_alunos_situacao_desvinculada', 'quantidade_matriculas_parfor',
    'quantidade_matriculas_financiamento_reembolsavel_fies',
    'quantidade_matriculas_financiamento_nao_reembolsavel_prouni_integral'
]

# 2. Processar por "Chunks" (Pedaços) para não estourar a RAM
caminho_base = '/content/base_unificada.csv'
chunks = []

print("Iniciando a filtragem por pedaços...")

# Lendo de 200.000 em 200.000 linhas
for chunk in pd.read_csv(caminho_base, usecols=colunas_foco, chunksize=200000, low_memory=False):
    # Filtrar apenas Licenciaturas em cada pedaço
    filtrado = chunk[chunk['tipo_grau_academico'] == 'Licenciatura'].copy()

    # Otimizar tipos numéricos (Downcasting) para economizar memória
    for col in filtrado.select_dtypes(include=['int64']).columns:
        filtrado[col] = pd.to_numeric(filtrado[col], downcast='integer')

    chunks.append(filtrado)

# 3. Unificar os pedaços filtrados
df_lic = pd.concat(chunks, ignore_index=True)
print(f"Filtragem concluída! Total de linhas de Licenciatura: {len(df_lic)}")

# 4. Criar as Variáveis Temporárias (Lags e Deltas)
df_lic = df_lic.sort_values(['id_curso', 'ano'])

# Criando a memória de 3 anos (Lag)
df_lic['matriculas_lag3'] = df_lic.groupby('id_curso')['quantidade_matriculas'].shift(3)

# Calculando a Tendência (O principal KPI da árvore)
df_lic['delta_matricula_3anos'] = (df_lic['quantidade_matriculas'] - df_lic['matriculas_lag3']) / df_lic['matriculas_lag3']

# 5. Criar a Target (Risco futuro - olhando 2 anos à frente)
df_lic['matriculas_futuro'] = df_lic.groupby('id_curso')['quantidade_matriculas'].shift(-2)
# Definimos Risco (1) se as matrículas caírem drasticamente ou o curso sumir (NaN)
df_lic['TARGET_RISCO'] = np.where((df_lic['matriculas_futuro'] < (df_lic['quantidade_matriculas'] * 0.5)) |
                                  (df_lic['matriculas_futuro'].isna()), 1, 0)

# Limpeza de valores infinitos ou NaNs gerados pelos cálculos
df_lic = df_lic.replace([np.inf, -np.inf], np.nan).dropna(subset=['delta_matricula_3anos', 'matriculas_futuro'])

# 6. Salvar em formato Parquet (Muito mais leve que CSV para a próxima vez)
df_lic.to_parquet('base_licenciaturas_modelagem.parquet')

print("Seu novo Dataset de modelagem foi salvo como 'base_licenciaturas_modelagem.parquet'!")
```

```python
import pandas as pd
import numpy as np

# 1. Carregar a base bruta das licenciaturas
df = pd.read_parquet('/content/passo1_licenciaturas_bruto.parquet')

# 2. Ordenação rigorosa (Essencial para o cálculo de tempo funcionar)
df = df.sort_values(['id_curso', 'ano'])

print("Iniciando transformações temporais...")

# 3. Criando as variáveis de "Retrovisor" (Lag e Delta de 3 anos)
# Precisamos saber como o curso estava 3 anos atrás para ver a tendência
df['matriculas_lag3'] = df.groupby('id_curso')['quantidade_matriculas'].shift(3)

# KPI 1: Variação de Matrículas em 3 anos (O motor da árvore)
df['delta_matricula_3anos'] = (df['quantidade_matriculas'] - df['matriculas_lag3']) / df['matriculas_lag3'].replace(0, np.nan)

# KPI 2: Taxa de Ocupação (Eficiência de Vagas)
df['taxa_ocupacao'] = df['quantidade_ingressantes'] / df['quantidade_vagas'].replace(0, np.nan)

# KPI 3: Índice de Evasão (Desvinculados / Matrículas)
df['indice_evasao'] = df['quantidade_alunos_situacao_desvinculada'] / df['quantidade_matriculas'].replace(0, np.nan)

# KPI 4: Dependência de Fomento (Fies + Prouni)
df['dep_fomento'] = (df['quantidade_matriculas_financiamento_reembolsavel_fies'] +
                     df['quantidade_matriculas_financiamento_nao_reembolsavel_prouni_integral']) / df['quantidade_matriculas'].replace(0, np.nan)

# 4. CRIANDO A TARGET (O Alvo: O curso entrou em colapso no futuro?)
# Vamos olhar 2 anos para o futuro (shift -2)
df['matriculas_futuro'] = df.groupby('id_curso')['quantidade_matriculas'].shift(-2)

# Definição de Risco (1): Se as matrículas caírem mais de 50% em 2 anos ou o curso sumir
df['TARGET_RISCO'] = np.where(
    (df['matriculas_futuro'] < (df['quantidade_matriculas'] * 0.5)) | (df['matriculas_futuro'].isna()),
    1, 0
)

# 5. Limpeza de Segurança
# Removemos apenas o que for essencialmente nulo para o modelo (o início e fim da série temporal)
df_final = df.dropna(subset=['delta_matricula_3anos', 'matriculas_futuro']).copy()

# Tratamento de valores infinitos que podem surgir em divisões
df_final = df_final.replace([np.inf, -np.inf], 0).fillna(0)

print(f"Transformação concluída! Sobraram {len(df_final)} linhas com histórico completo para modelagem.")

# 6. Salvar o Data-mart Final
df_final.to_parquet('/content/passo2_licenciaturas_modelagem.parquet')
print("Arquivo final salvo: /content/passo2_licenciaturas_modelagem.parquet")

# Visualização rápida do resultado
display(df_final[['ano', 'nome_area_detalhada', 'delta_matricula_3anos', 'TARGET_RISCO']].head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. CARREGAR OS DADOS
df_model = pd.read_parquet('/content/passo2_licenciaturas_modelagem.parquet')

# 2. DEFINIR VARIÁVEIS (X = Preditores, y = Alvo)
# Estas são as 4 colunas que você criou na engenharia de atributos
features = ['delta_matricula_3anos', 'taxa_ocupacao', 'indice_evasao', 'dep_fomento']
X = df_model[features]
y = df_model['TARGET_RISCO']

# 3. DIVISÃO TRIPARTIDA (60% Treino, 20% Validação, 20% Teste)
# Primeiro, isolamos o Teste (20%) - que é a "prova final"
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Dos 80% que sobraram, dividimos em Treino (75% de 80 = 60%) e Validação (25% de 80 = 20%)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42)

print(f"Estrutura dos Dados:")
print(f"- Treino: {len(X_train)} registros")
print(f"- Validação: {len(X_val)} registros")
print(f"- Teste: {len(X_test)} registros\n")

# 4. TREINAMENTO DA ÁRVORE (Parametrizada para explicabilidade)
# max_depth=3 para o professor conseguir ler cada nó sem esforço
modelo = DecisionTreeClassifier(max_depth=3, criterion='gini', random_state=42)
modelo.fit(X_train, y_train)

# 5. AVALIAÇÃO (Acurácia na base de Validação)
y_pred_val = modelo.predict(X_val)
acuracia = accuracy_score(y_val, y_pred_val)

print("--- RESULTADOS DA MODELAGEM ---")
print(f"Acurácia Global: {acuracia:.2%}")
print("\nRelatório de Métricas (Precisão e Recall):")
print(classification_report(y_val, y_pred_val))

# 6. VISUALIZAÇÃO DA ÁRVORE (A "Folhagem" Aberta)
plt.figure(figsize=(25,12))
plot_tree(modelo,
          feature_names=features,
          class_names=['Estável (0)', 'Risco (1)'],
          filled=True,
          rounded=True,
          proportion=True, # Mostra o % de amostras em cada folha
          precision=2,
          fontsize=12)

plt.title("Árvore de Decisão: Diagnóstico de Risco de Colapso (Licenciaturas)", fontsize=16)
plt.show()

# 7. IMPORTÂNCIA DAS VARIÁVEIS
importancias = pd.Series(modelo.feature_importances_, index=features).sort_values(ascending=False)
print("\n--- O que mais influenciou o risco? ---")
print(importancias)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

# Calcular a probabilidade (necessário para a ROC)
y_probs = modelo.predict_proba(X_val)[:, 1]
auc_valor = roc_auc_score(y_val, y_probs)

print(f"ROC-AUC: {auc_valor:.2f}")

# Gerar o gráfico da Curva ROC
fpr, tpr, thresholds = roc_curve(y_val, y_probs)
plt.plot(fpr, tpr, label=f'Modelo (AUC = {auc_valor:.2f})')
plt.plot([0, 1], [0, 1], 'k--') # Linha de base (aleatória)
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos (Recall)')
plt.title('Curva ROC-AUC')
plt.legend()
plt.show()
```

```python
# Gerando as probabilidades de risco para os dados
probalidades = modelo.predict_proba(X)[:, 1]

# Criando uma nova coluna no DataFrame original para o Gestor ver
df_model['SCORE_RISCO'] = probalidades

# Definindo as faixas de risco
def classificar_severidade(p):
    if p <= 0.20: return 'Baixo Risco'
    elif p <= 0.60: return 'Risco Moderado'
    else: return 'Alto Risco'

df_model['STATUS_ALERTA'] = df_model['SCORE_RISCO'].apply(classificar_severidade)

# Verificando quantos cursos caíram em cada faixa
print(df_model['STATUS_ALERTA'].value_counts())
```

```python
# 1. Filtrar o último ano disponível
ultimo_ano = df_model['ano'].max()
df_atual = df_model[df_model['ano'] == ultimo_ano].copy()

# 2. Filtrar apenas quem o modelo classificou como 'Alto Risco'
lista_negra = df_atual[df_atual['STATUS_ALERTA'] == 'Alto Risco']

# 3. Criar a tabela com os nomes das áreas (cursos) e os indicadores que causaram o risco
# Note que usei 'nome_area_detalhada' que é o nome que apareceu na sua lista
resultado_final = lista_negra[[
    'id_curso',
    'nome_area_detalhada',
    'delta_matricula_3anos',
    'SCORE_RISCO'
]].sort_values(by='SCORE_RISCO', ascending=False)

print(f"--- Top 20 Áreas/Cursos em Alto Risco em {ultimo_ano} ---")
print(resultado_final.head(20))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrando o último ano para o gráfico não ficar poluído
ultimo_ano = df_model['ano'].max()
df_plot = df_model[df_model['ano'] == ultimo_ano].copy()

# Definindo o tamanho das bolhas baseado nas matrículas
# (ajustamos a escala para as bolhas não ficarem gigantes nem minúsculas)
tamanho_bolhas = (df_plot['quantidade_matriculas'] / df_plot['quantidade_matriculas'].max()) * 1000 + 50

plt.figure(figsize=(14, 8))

# Criando o gráfico de dispersão (bolhas)
scatter = plt.scatter(
    x=df_plot['delta_matricula_3anos'],
    y=df_plot['SCORE_RISCO'],
    s=tamanho_bolhas,
    c=df_plot['STATUS_ALERTA'].map({'Baixo Risco': 'green', 'Risco Moderado': 'orange', 'Alto Risco': 'red'}),
    alpha=0.5,
    edgecolors='w'
)

# Linhas de referência para ajudar a explicação
plt.axhline(y=0.6, color='red', linestyle='--', alpha=0.3, label='Zona de Perigo (Score > 0.6)')
plt.axvline(x=1.0, color='blue', linestyle='--', alpha=0.3, label='Expansão Crítica (> 100%)')

plt.title(f'Mapa de Vulnerabilidade das Licenciaturas ({ultimo_ano})', fontsize=16, fontweight='bold')
plt.xlabel('Taxa de Expansão (Delta de Matrículas em 3 anos)', fontsize=12)
plt.ylabel('Probabilidade de Colapso (Score de Risco)', fontsize=12)

# Adicionando uma legenda manual para as cores
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Baixo Risco', markerfacecolor='green', markersize=15),
    Line2D([0], [0], marker='o', color='w', label='Risco Moderado', markerfacecolor='orange', markersize=15),
    Line2D([0], [0], marker='o', color='w', label='Alto Risco', markerfacecolor='red', markersize=15)
]
plt.legend(handles=legend_elements, title="Status de Alerta", loc='upper left')

plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
