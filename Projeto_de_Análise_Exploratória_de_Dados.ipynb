# Análise Exploratória de Dados (EDA) em Python

Para realizar uma análise exploratória dos conjuntos de dados fornecidos, é necessário primeiro carregar os dados e explorar sua estrutura. A seguir, apresento um código Python que realiza a EDA, respondendo a algumas perguntas relevantes.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregamento dos dados
energy_types = pd.read_csv('/content/energy_types.csv')
country_totals = pd.read_csv('/content/country_totals.csv')
```

# 1. Análise de Distribuição de Dados

- Análise

Este bloco de código imprime as estatísticas descritivas para o DataFrame country_totals. Novamente, são exibidas as estatísticas descritivas para cada coluna numérica no DataFrame.
Aplicação prática

Essas estatísticas descritivas ajudam a compreender a distribuição e as características dos dados em country_totals.
Em resumo, os blocos de código fornecem informações valiosas sobre a natureza dos dados em seus respectivos DataFrames, facilitando a análise exploratória e a compreensão inicial das características dos conjuntos de dados.

```python
# Estatísticas descritivas
print("Estatísticas Descritivas de Energy Types:")
print(energy_types.describe())

print("\nEstatísticas Descritivas de Country Totals:")
print(country_totals.describe())
```

# 2. Transformações dos Dados
- Análise

Neste bloco de código, está sendo realizada a normalização dos dados na coluna '2016' do DataFrame country_totals. A normalização é feita dividindo cada valor pelo somatório total da coluna '2016'.
- Aplicação prática

A normalização é útil quando se deseja expressar os valores em uma escala relativa, em vez de absoluta. No contexto de produção de energia, normalizar os dados pelo total pode ajudar a comparar a contribuição de cada país de uma maneira proporcional ao total global, facilitando a análise das proporções de produção de energia em relação ao todo.
Este tipo de operação é frequentemente útil em análises comparativas, onde você deseja avaliar o peso relativo de cada país na produção total de energia.

```python
# Normalização dos dados (por exemplo, normalizar a produção de energia pelo total)
country_totals['2016_normalized'] = country_totals['2016'] / country_totals['2016'].sum()
```

# 3. Análise de Tabelas de Contingência ou Agregações

- Análise

Este bloco de código utiliza a função pd.crosstab para criar uma tabela de contingência, que mostra a contagem de observações para combinações específicas de categorias. Neste caso, a tabela representa a distribuição dos tipos de energia por país em 2018.
- Aplicação prática

Tabelas de contingência são frequentemente usadas para entender a distribuição conjunta de duas variáveis categóricas. No contexto de energia, esta tabela pode fornecer insights sobre quais tipos de energia são mais prevalentes em diferentes países, facilitando a identificação de padrões ou discrepâncias na matriz energética global.
Essa análise pode ser valiosa para formuladores de políticas energéticas, empresas do setor e pesquisadores que buscam compreender as preferências e diversidade nas fontes de energia utilizadas pelos países em um determinado ano.

```python
# Tabela de contingência para tipo de energia por país em 2018
contingency_table = pd.crosstab(energy_types['country_name'], energy_types['type'], margins=True, margins_name='Total')
print("Tabela de Contingência - Tipo de Energia por País (2018):")
print(contingency_table)
```

# 4. Análise de Correlação
- Análise

Este trecho calcula a matriz de correlação entre os anos de produção total de energia por país (nos anos de 2016, 2017 e 2018) e a visualiza usando um mapa de calor.
- Aplicação prática

A correlação entre os anos de produção de energia pode fornecer insights sobre padrões temporais na produção de energia de um país. Por exemplo, uma alta correlação positiva entre dois anos pode indicar consistência na produção ao longo do tempo, enquanto uma correlação negativa pode sugerir variações sazonais ou mudanças significativas na produção de energia.
Essa análise é relevante para entender as tendências de produção de energia ao longo do tempo e pode ser útil para prever a demanda futura, identificar necessidades de infraestrutura e tomar decisões informadas sobre políticas energéticas.

```python
# Correlação entre os anos de produção de energia total por país
correlation_matrix = country_totals[['2016', '2017', '2018']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlação entre os Anos de Produção de Energia Total por País')
plt.show()
```

# 5. Análise de Informação Mútua ou Atributos Derivados
- Análise

Este código utiliza a função mutual_info_score do scikit-learn para calcular a informação mútua entre o tipo de energia e o país.
A informação mútua é calculada para cada tipo de energia, e os resultados são visualizados em um gráfico de barras.
- Aplicação prática

A informação mútua entre o tipo de energia e o país pode ajudar a entender o quão informativo é o tipo de energia para prever o país de origem. Essa análise é útil em tarefas de classificação, onde o objetivo é prever o país com base no tipo de energia. O gráfico de barras fornece uma representação visual dessas relações informativas. Essas informações podem ser valiosas para otimizar modelos de machine learning e melhorar a precisão das previsões.

```python
from sklearn.metrics import mutual_info_score

# Calcular informação mútua
mi_scores = []

for col in energy_types['type'].unique():
    mask = (energy_types['type'] == col)
    mi_scores.append(mutual_info_score(energy_types['country_label'][mask], energy_types['type_encoded'][mask]))

# Criar DataFrame com resultados
mi_df = pd.DataFrame({'Type': energy_types['type'].unique(), 'Mutual_Information': mi_scores})

# Visualizar informação mútua
sns.barplot(x='Mutual_Information', y='Type', data=mi_df, orient='h')
plt.title('Informação Mútua entre Tipo de Energia e País')
plt.show()
```
